```python
import torch
from torch import nn
from PIL import Image
import torchvision
from torch.utils.data import Dataset, DataLoader
import os
import pandas as pd
import numpy as np
from d2l import torch as d2l
```

```python
# 所有的类别名称
leaves_classes = sorted(list(set(labels_df['label'])))

# 所有的类别数
num_classes = len(leaves_classes)
print(f'训练样本：{len(labels_df)}')
print(f'类别：{num_classes}')
```

```python
# 建立类别名称到类别索引的映射
class_to_idx = dict(zip(leaves_classes, range(num_classes)))
```

## 自定义Dataset

```python
class LeavesData(Dataset):
    def __init__(self, root_dir, csv_path, transform, mode='train', valid_ratio=0.2):
        """
        Args:
            root_dir: 数据集的根目录
            csv_path: csv文件路径
            transform: 数据增广
            mode: 'train'->训练集，'valid'->验证集，'test'->测试集, 'train_valid'->竞赛给出的全部训练集
            valid_ratio: 验证集占训练集的比重
        """
        self.root_dir = root_dir
        self.mode = mode
        # 读取csv文件
        self.data_info = pd.read_csv(csv_path)
        # 数据集长度
        self.data_len = len(self.data_info)
        # 测试集不需要拆分
        if mode != 'test': 
            # 拆分训练集和验证集
            self.train_len = int(self.data_len * (1 - valid_ratio)) 
        # 数据增广
        self.augs = transform
 
        if mode == 'train':
            self.image_names = np.asarray(self.data_info.iloc[:self.train_len, 0])
            self.image_labels = np.asarray(self.data_info.iloc[:self.train_len, 1])
        elif mode == 'valid':
            self.image_names = np.asarray(self.data_info.iloc[self.train_len:, 0])
            self.image_labels = np.asarray(self.data_info.iloc[self.train_len:, 1])
        # 测试集的标签是未知的
        elif mode == 'test':
            self.image_names = np.asarray(self.data_info.iloc[:, 0])
            self.image_labels = np.asarray(['unknown'] * len(self.image_names))
        # 全部的训练集，调好超参数后在这个训练集重新训练一个模型
        elif mode == 'train_valid':
            self.image_names = np.asarray(self.data_info.iloc[:self.data_len, 0])
            self.image_labels = np.asarray(self.data_info.iloc[:self.data_len, 1])
        
        print(f'Finished reading the {mode} set of Leaves Dataset({len(self.image_names)} samples found)')
   
    def __getitem__(self, index):
        # 读取图片
        img = Image.open(os.path.join(self.root_dir, self.image_names[index]))
        img = self.augs(img)
        label_idx = class_to_idx[self.image_labels[index]] if self.image_labels[index] != 'unknown' else -1
        return img, label_idx
        
    def __len__(self):
        return len(self.image_names)
```

## 数据增广

```python
train_valid_augs = torchvision.transforms.Compose([
    torchvision.transforms.RandomHorizontalFlip(p=0.5),  
    torchvision.transforms.RandomVerticalFlip(p=0.5),  
    torchvision.transforms.ToTensor()
])

test_augs = torchvision.transforms.Compose([
    torchvision.transforms.ToTensor()
])
```

## 加载数据集

```python
root_dir = '../data/classify-leaves'
train_valid_csv_path = os.path.join(root_dir, 'train.csv')
test_csv_path = os.path.join(root_dir, 'test.csv')

train_ds = LeavesData(root_dir, train_valid_csv_path, train_valid_augs, mode='train')
valid_ds = LeavesData(root_dir, train_valid_csv_path, train_valid_augs, mode='valid')
test_ds = LeavesData(root_dir, test_csv_path, test_augs, mode='test')
train_valid_ds = LeavesData(root_dir, train_valid_csv_path, train_valid_augs, mode='train_valid')
```

```python
batch_size = 128
train_iter = torch.utils.data.DataLoader(
        dataset=train_ds, batch_size=batch_size,
        shuffle=True, drop_last=True
    )

valid_iter = torch.utils.data.DataLoader(
        dataset=valid_ds, batch_size=batch_size,
        shuffle=True, drop_last=True
    )

test_iter = torch.utils.data.DataLoader(
        dataset=test_ds, batch_size=batch_size,
        shuffle=False, drop_last=False
    )

train_valid_iter = torch.utils.data.DataLoader(
        dataset=train_valid_ds, batch_size=batch_size,
        shuffle=True, drop_last=True
    )
```

## 微调resnet18模型
通过训练集和验证集来调超参数

```python
def get_net(num_classes, pretrained=True):
    net = torchvision.models.resnet18(pretrained=pretrained)
    net.fc = nn.Linear(net.fc.in_features, num_classes)
    nn.init.xavier_uniform_(net.fc.weight)
    return net
```

```python
# 如果param_group=True，输出层中的模型参数将使用十倍的学习率
def train(net, train_iter, valid_iter, learning_rate, wd, num_epochs, param_group=True):
    """进行训练模型"""
    devices = d2l.try_all_gpus()
    loss = nn.CrossEntropyLoss(reduction="none")
    
    if param_group:
        # 输出层使用较大学习率
        params_1x = [param for name, param in net.named_parameters() if name not in ["fc.weight", "fc.bias"]]
        # 使用字典列表的形式，对不同层设置不同的学习率
        trainer = torch.optim.SGD([{'params': params_1x}, {'params': net.fc.parameters(), 'lr': learning_rate * 10}],
                                    lr=learning_rate, weight_decay=wd)
    else:
        trainer = torch.optim.SGD(net.parameters(), lr=learning_rate, weight_decay=wd)
    
    num_batches, timer = len(train_iter), d2l.Timer()
    legend = ['train loss', 'train acc']
    
    if valid_iter is not None:
        legend.append('valid acc')
    animator = d2l.Animator(xlabel='epoch', xlim=[1, num_epochs], legend=legend)
    net = nn.DataParallel(net, device_ids=devices).to(devices[0])
    
    for epoch in range(num_epochs):
        net.train()
        metric = d2l.Accumulator(3) # 损失，预测正确个数，样本数
        # --------Train----------
        for i, (features, labels) in enumerate(train_iter):
            timer.start()
            l, acc = d2l.train_batch_ch13(net, features, labels, loss, trainer, devices)
            metric.add(l, acc, labels.shape[0])
            timer.stop()
            # 满足下面条件就进行画图展示
            if (i + 1) % (num_batches // 5) == 0 or i == num_batches - 1:
                animator.add(epoch + (i + 1) / num_batches,
                             (metric[0] / metric[2], # 平均损失
                              metric[1] / metric[2], # 精度
                              None))
        # --------Valid----------
        if valid_iter is not None:
            valid_acc = d2l.evaluate_accuracy_gpu(net, valid_iter)
            animator.add(epoch + 1, (None, None, valid_acc))       
    
    measures = (f'train loss {metric[0] / metric[2]:.3f}',
                f'train acc {metric[1] / metric[2]:.3f}')
    
    if valid_iter is not None:
        measures += (f', valid acc {valid_acc:.3f}',)
    
    print(*measures)
    print(f'{metric[2] * num_epochs / timer.sum():.1f} examples/sec on {str(devices)}')
```

```python
# 超参数
learning_rate = 3e-4
weight_decay = 1e-3
num_epochs = 10
net = get_net(num_classes).to(d2l.try_gpu())
```

```python
train(net, train_iter, valid_iter, learning_rate, weight_decay, num_epochs)
```

## 重新训练模型
当我们对超参数满意之后，在本次竞赛的全部训练集上重新训练模型

```python
net= get_net(num_classes)
train(net, train_valid_iter, None, learning_rate, weight_decay, num_epochs)
```

```python
# 保存模型参数
torch.save(net.state_dict(),'resnet18_LeavesData_0.982.params')
```

## 在测试集上进行预测并上传
对测试集进行预测，保存结果（这里我gpu内存不够，我重新读取模型参数再预测的）

```python
# 加载模型
trained_net = get_net(num_classes)
trained_net.load_state_dict(torch.load('resnet18_LeavesData_0.982.params'))
trained_net.eval()
trained_net = trained_net.to(torch.device('cuda:0'))

preds = []
for X, _ in test_iter:
    y_hat = trained_net(X.to(torch.device('cuda:0')))
    # 保存预测的类别索引
    preds.extend(y_hat.argmax(dim=1).type(torch.int32).cpu().numpy())
    
test_df = pd.read_csv(test_csv_path)
submission_df = pd.DataFrame({'image':test_df['image'],'label':preds})
# 类别索引转换成类别名称
submission_df['label'] = submission_df['label'].apply(lambda x: leaves_classes[x])
submission_df.to_csv('submission.csv', index=False)
```
