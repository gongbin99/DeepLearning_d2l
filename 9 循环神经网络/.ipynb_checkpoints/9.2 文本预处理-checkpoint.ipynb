## 文本预处理：文本读进来变换为整型向量
常见的文本预处理步骤：


```python
import collections
import re
from d2l import torch as d2l
```

## 读入数据集
我们从H.G.Well的时光机器中加载文本。 这是一个相当小的语料库，只有30000多个单词，但足够我们小试牛刀， 而现实中的文档集合可能会包含数十亿个单词。 

```python
d2l.DATA_HUB['time_machine'] = (d2l.DATA_URL + 'timemachine.txt',
                                '090b5e7e70c295757f55df93cb0a180b9691891a')

def read_time_machine():
    """将时间机器数据集加载到文本行的列表中"""
    with open(d2l.download('time_machine'), 'r') as f:
        lines = f.readlines() # lines是字符串列表，每个元素是每行文本的字符串
    
    # 将文本中不是英文字母的字符全部替换成' '，再消除'\n'，再转为小写字母
    # 一般并不会这么做
    return [re.sub('[^A-Za-z]+', ' ', line).strip().lower() for line in lines]
```

```python
lines = read_time_machine()
print(f'文本总行数: {len(lines)}')
# 打印两行看看结果
print(lines[0])
print(lines[10])
```

## 词元化
将文本行拆成词元(token)列表，词元作为文本行的基本单位

```python
def tokenize(lines, token='word'):
    if token == 'word':
        # 按空格拆分为单词列表
        return [line.split() for line in lines]
    elif token == 'char':
        # 字符串变为字符列表
        return [list(line) for line in lines]
    else:
        print(f'错误: 未知词元类型: {token}')

tokens = tokenize(lines, token='word')
for i in range(10):
    print(tokens[i])
```

## 词表
对唯一词元进行统计，统计结果称为语料(corpus)，再将语料按出现频率从高到低排序，为每个预料分配一个索引。很少出现的词元通常被移除，这可以降低复杂性。 另外，语料库中不存在或已删除的任何词元都将映射到一个特定的未知词元“unk”。

我们可以选择增加一个列表，用于保存那些被保留的词元， 例如：填充词元（“pad”）； 序列开始词元（“bos”）； 序列结束词元（“eos”）。

```python
class Vocab: 
    """文本词表, 关键是实现词元到索引、索引到词元的转换"""
    def __init__(self, tokens=None, min_freq=0, reserved_tokens=None):
        if tokens is None:
            tokens = []
        if reserved_tokens is None:
            reserved_tokens = []
        # 统计每个语料的出现频率
        counter = count_corpus(tokens)
        # 将语料按出现频率从高到低排序
        self._token_freqs = sorted(counter.items(), 
                                   key=lambda x: x[1],
                                   reverse=True)
        # 未知词元的索引为0
        # 被保留的词元先保存
        self.idx_to_token = ['<unk>'] + reserved_tokens
        # 通过语料获得对应的索引
        self.token_to_idx = {token: idx
                             for idx, token in enumerate(self.idx_to_token)}
        
        for token, freq in self._token_freqs:
            if freq < min_freq:
                break
            # 插入从数据集中获得的语料
            if token not in self.token_to_idx:
                self.idx_to_token.append(token)
                self.token_to_idx[token] = len(self.idx_to_token) - 1

    def __len__(self):
        """词表的长度"""
        return len(self.idx_to_token)

    def __getitem__(self, tokens):
        """获取token对应的索引"""
        # 只查一个词元对应的索引
        if not isinstance(tokens, (list, tuple)):
            return self.token_to_idx.get(tokens, self.unk)
        # 查很多词元的索引
        return [self.__getitem__(token) for token in tokens]

    def to_tokens(self, indices):
        """通过索引获取相应的词元"""
        if not isinstance(indices, (list, tuple)):
            return self.idx_to_token[indices]
        return [self.idx_to_token[index] for index in indices]

    @property
    def unk(self):  # 未知词元的索引为0
        return 0

    @property
    def token_freqs(self):
        return self._token_freqs

def count_corpus(tokens):
    """统计语料的频率"""
    # 这里的tokens是1D列表或2D列表
    if len(tokens) == 0 or isinstance(tokens[0], list):
        # 将所有词元构成一个列表
        tokens = [token for line in tokens for token in line]
    
    # 获得每个语料到出现次数的映射
    return collections.Counter(tokens)
```

```python
vocab = Vocab(tokens)
list(vocab.token_to_idx.items())[:10]
```

现在，能够将每一条文本转换成一个数字索引表

```python
for i in [0, 10]:
    print(f'文本: {tokens[i]}')
    print(f'索引: {vocab[tokens[i]]}')
```

## 整合上面所有功能函数

```python
def load_corpus_time_machine(max_tokens=-1):
    """返回时光机器数据集的词元索引列表和词表"""
    # 按行读入文本
    lines = read_time_machine()
    # 词元化
    tokens = tokenize(lines, 'char')
    # 构建词表
    vocab = Vocab(tokens)
    # 因为时光机器数据集中的每个文本行不一定是一个句子或一个段落，
    # 所以将所有文本行展平到一个列表中
    corpus = [vocab[token] for line in tokens for token in line]
    if max_tokens > 0:
        corpus = corpus[:max_tokens]
    return corpus, vocab
```

```python
corpus, vocab = load_corpus_time_machine()
len(corpus), len(vocab)
```
