#### 思考数据适合用什么类型的图来表示很重要

## 饼图
根据百分比展示的数据

```python
import matplotlib.pyplot as plt
```

```python
labels = 'Frogs', 'Hogs', 'Dogs', 'Logs'
# 每个标签对应的比例
sizes = [15, 30, 45, 10]
# 突出第二个标签
explode = (0, 0.1, 0, 0)
plt.pie(sizes, # 各个标签所占比例
        explode=explode,  # 是否突出某个标签
        labels=labels,  # 所有的标签
        autopct='%.1f%%',  # 显示百分数的方式
        shadow=False,  # 不带阴影就是二维饼图
        startangle=90) # 饼图起始的角度
plt.axis('equal')
plt.show()
```

## 直方图
展示数据的分布

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(0)
mu, sigma = 100, 20 #均值和标准差
a = np.random.normal(mu, sigma, size=100)

# 每个直方中包含了多少个原来数组a的元素，构成了直方的高度
plt.hist(a, 
         20,  # bins: 直方的个数
         histtype='stepfilled', 
         facecolor='green',
         alpha=0.75)
plt.title('Histogram')

plt.show()
```

## 极坐标图
在角度空间内展示数据，用的不多


```python
N = 20  # 数据的个数
theta = np.linspace(0.0, 2 * np.pi,  # 0 - 360度
                    N, endpoint=False) # 等分出N个角度
radii = 10 * np.random.rand(N) # 每个角度对应的值
width = np.pi / 4 * np.random.rand(N) # 宽度值

ax = plt.subplot(111, 
                 projection='polar') # 表名绘制极坐标图
bars = ax.bar(theta,  # left 绘制的角度位置
              radii,  # height 中心点往边缘绘制的长度r
              width=width,  # width：区域的面积 
              bottom=0.0)

for r, bar in zip(radii, bars):
    bar.set_facecolor(plt.cm.viridis(r / 10.))
    bar.set_alpha(0.5)
    
plt.show()
```

## 散点图


```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 默认绘制区域111，当前绘制区域的对象是ax（推荐使用面向对象的方式绘图）
fig, ax = plt.subplots()
# randn是随机生成符合正态分布的函数
ax.plot(10 * np.random.randn(100), 10 * np.random.randn(100), 'o')
ax.set_title('Simple Scatter')

plt.show()
```
