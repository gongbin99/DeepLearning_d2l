query就是你想要干嘛（主观意志），比如说我想要喝咖啡，那么环境等因素就是key-value对（key和value可以相同可以不同），咖啡杯摆在那里可以作为key，value可以是这个咖啡杯对我的价值


随意线索：想要干啥，所以会关注的事情，比如想和咖啡，那么便会关注咖啡杯<br>
不随意线索：与想干的事情无关，比如我想喝咖啡，那么无需关注报纸

注意力机制 (attention pooling)：根据query有偏向的选择一些东西（key-value pair）

## 非参注意力池化层
这里的数据就是(key, value)对<br>
平均池化可以视为它的特例，不关心query，就把所有的value求个均值<br>
K(x)是衡量x和x_i距离的函数，y_i前面的项视为相对重要性，加权对y_i求和。<br>
<b>假设给出x，会选择跟我的x比较接近的x_i，看它的y_i</b>，这里的Nadaraya-Watson是非参的不需要学任何东西，来个新值x就用所有数据去查，它与现在的注意力池化层是很相近的
