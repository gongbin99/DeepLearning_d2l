
输入的query与所有训练样本的key计算出注意力分数，将注意力分数向量经过softmax得到注意力权重（均大于0，和为1），最后输出所有训练样本value的加权和

## 将q, k, v从标量扩展到高维情况

注意力权重的计算就是对注意力分数进行softmax，a是计算注意力分数的函数，下面讲解如何设计a

## 两种常见的计算注意力分数的方式

### a的是设计方式一：Additive Attention
<b>好处：key、value、query可以是任意长度</b>


### a的设计方式二：Scaled Dot-Product Attention
<b>前提：需要query和key的长度相同</b><br>
<b>好处：无需训练参数</b>


## 总结


```python
import math
import torch
from torch import nn
from d2l import torch as d2l
```

## 前菜：掩蔽softmax操作的实现
softmax作用于query和key的注意力分数上，得到注意力权重。在某些情况下，并非所有值都应该被我们注意。例如：我们对文本序列填充了无意义的特殊词元，为了只对有意义的词元作为值进行注意力汇聚，我们可以指定一个有效序列长度（即词元的个数），以便在计算softmax时过滤掉超出指定范围的值。

```python
def masked_softmax(X, valid_lens):
    # X:3D张量，valid_lens:1D或2D张量
    if valid_lens is None:
        return nn.functional.softmax(X, dim=-1)
    else:
        # 需要掩蔽
        shape = X.shape
        # 如果是一维的张量
        if valid_lens.dim() == 1:
            valid_lens = torch.repeat_interleave(valid_lens, shape[1])
        else:
            # 如果是二维，拉成一维的
            valid_lens = valid_lens.reshape(-1)
            
        # 最后一轴上被掩蔽的元素使用一个非常大的负值替换，从而其softmax输出为0
        X = d2l.sequence_mask(X.reshape(-1, shape[-1]), valid_lens,
                              value=-1e6)
        return nn.functional.softmax(X.reshape(shape), dim=-1)
```

```python
# 测试：对于两个2x4的矩阵，合理的长度分别为2和3
masked_softmax(torch.rand(2, 2, 4), torch.tensor([2, 3]))
```

```python
# 测试：同样，也可以使用二维张量表示Valid_lens，为矩阵样本中的每一行指定有效长度。
masked_softmax(torch.rand(2, 2, 4), torch.tensor([[1, 3], [2, 4]]))
```

## Additive Attention的实现

将query和key连结起来后输入到一个多层感知机（MLP）中, 感知机包含一个隐藏层，其隐藏单元数是一个超参数h。 通过使用tanh作为激活函数，并且禁用偏置项。

```python
class AdditiveAttention(nn.Module):
    """加性注意力"""
    def __init__(self, key_size, query_size, 
                 num_hiddens, dropout, **kwargs):
        super(AdditiveAttention, self).__init__(**kwargs)
        # 看上述公式可知，无需bias
        # 将key转换成h维
        self.W_k = nn.Linear(key_size, num_hiddens, bias=False)
        # 将query转换成h维
        self.W_q = nn.Linear(query_size, num_hiddens, bias=False)
        self.w_v = nn.Linear(num_hiddens, 1, bias=False)
        self.dropout = nn.Dropout(dropout)

    def forward(self, queries, keys, values, valid_lens):
        # 经过全连接层后的query和key
        queries, keys = self.W_q(queries), self.W_k(keys)
        # 在维度扩展后，
        # queries的形状：(batch_size，查询的个数，1，num_hidden)
        # key的形状：(batch_size，1，“键－值”对的个数，num_hiddens)
        # 使用广播方式进行求和
        features = queries.unsqueeze(2) + keys.unsqueeze(1)
        features = torch.tanh(features)
        
        # self.w_v仅有一个输出，因此从形状中移除最后那个维度。
        # scores的形状：(batch_size，查询的个数，“键-值”对的个数)
        scores = self.w_v(features).squeeze(-1)
        # 得到注意力权重
        self.attention_weights = masked_softmax(scores, valid_lens)
        # values的形状：(batch_size，“键－值”对的个数，值的维度)
        return torch.bmm(self.dropout(self.attention_weights), values)
```

演示上面的AdditiveAttention类， 其中查询、键和值的形状为（批量大小，步数或词元序列长度，特征维度），query和key的特征维度不同

```python
# queries有两个批量，1个20维的query
# keys有两个批量，10个2维的key
queries, keys = torch.normal(0, 1, (2, 1, 20)), torch.ones((2, 10, 2))
# 两个批量的value矩阵是相同的，每个批量有10个4维的value（个数与key一样）
values = torch.arange(40, dtype=torch.float32).reshape(1, 10, 4).repeat(2, 1, 1)

# 两个文本序列的有效长度
valid_lens = torch.tensor([2, 6])

# 加性注意力汇聚层
attention = AdditiveAttention(key_size=2, 
                              query_size=20, 
                              num_hiddens=8, # 超参数
                              dropout=0.1)
attention.eval()
attention(queries, keys, values, valid_lens)
```

尽管加性注意力包含了可学习的参数，但由于本例子中每个键都是相同的， 所以注意力权重是均匀的。（两个query的长度一个2，一个6）

```python
d2l.show_heatmaps(attention.attention_weights.reshape((1, 1, 2, 10)),
                  xlabel='Keys', ylabel='Queries')
```

## 缩放点积注意力
使用点积可以得到计算效率更高的评分函数， 但是点积操作要求查询和键具有相同的长度d



```python
# 下面的缩放点积注意力的实现使用了暂退法进行模型正则化
class DotProductAttention(nn.Module):
    """缩放点积注意力"""
    def __init__(self, dropout, **kwargs):
        super(DotProductAttention, self).__init__(**kwargs)
        self.dropout = nn.Dropout(dropout)

    # queries的形状：(batch_size，查询的个数，d)
    # keys的形状：(batch_size，“键－值”对的个数，d)
    # values的形状：(batch_size，“键－值”对的个数，value的维度)
    # valid_lens的形状:(batch_size，)或者(batch_size，查询的个数)  ps: 后者可以指定每个查询的有效长度
    def forward(self, queries, keys, values, valid_lens=None):
        d = queries.shape[-1]
        # 交换keys的最后两个维度与queries进行批量矩阵相乘，再除以√d，得到每个查询对每个value的注意力分数
        scores = torch.bmm(queries, keys.transpose(1,2)) / math.sqrt(d)
        # 求出注意力权重
        self.attention_weights = masked_softmax(scores, valid_lens)
        # 输出每个query对values的加权和
        return torch.bmm(self.dropout(self.attention_weights), values)
```

```python
queries = torch.normal(0, 1, (2, 1, 2))
attention = DotProductAttention(dropout=0.5)
attention.eval()
attention(queries, keys, values, valid_lens)
```

```python
d2l.show_heatmaps(attention.attention_weights.reshape((1, 1, 2, 10)),
                  xlabel='Keys', ylabel='Queries')
```
