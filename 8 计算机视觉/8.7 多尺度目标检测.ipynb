## 多尺度锚框
若对原图每个像素生成5个锚框，原图分辨率561 x 728，那么将在这一张图上标记和预测超过200w个锚框（561 x 728 x 5）。因此我们需要减少图像上的锚框数量。

在不同尺度下，我们可以生成不同数量和不同大小的锚框。比起较大的目标，较小的目标在图像上出现的可能性更多样。例如：1x1、1x2、2x2的目标能够以4、2、1种方式出现在2x2的图像上。 因此，为了检测原图中较小的物体时，我们可以在较大的区域(特征图的尺寸)上使用较小的锚框，而对于检测较大的物体，我们可以在较少的区域，使用较大的锚框。

### 演示在多尺度下生成锚框

```python
%matplotlib inline
import torch
from d2l import torch as d2l

img = d2l.plt.imread('../img/catdog.jpg')
h, w = img.shape[:2]
h, w
```

特征图feature map: 我们将卷积层输出的二维数组称为特征图

由于锚框中的(x, y)轴坐标值已经除以特征图（fmap）的宽度和高度，因此这些值介于0和1之间，表示特征图中锚框的相对位置。将这些相对坐标 * 原图的高宽 = 锚框在原图的坐标

```python
def display_anchors(fmap_w, fmap_h, s):
    """将在特征图上生成的若干锚框放在原图上显示"""
    d2l.set_figsize()
    # 前两个维度上的值不影响输出
    fmap = torch.zeros((1, 10, fmap_h, fmap_w))
    # anchors.shape (批量大小，锚框数量，4)
    anchors = d2l.multibox_prior(fmap,  
                                 sizes=s,  # 锚框面积/原图面积=s^2
                                 ratios=[1, 2, 0.5]) # 锚框的宽/高
    
    bbox_scale = torch.tensor((w, h, w, h)) # 原图的宽高
    
    # 将特征图上生成的锚框在原图上显示
    d2l.show_bboxes(d2l.plt.imshow(img).axes,
                    anchors[0] * bbox_scale) # 相对坐标转换为原图中的绝对坐标
```

在较大区域（fmap = 4 x 4）使用较小的锚框，能够采样原图更多的区域，便于检测小物体

```python
display_anchors(fmap_w=4, fmap_h=4, s=[0.15])
```

在较小的区域（fmap = 2 x 2），使用较大锚框，能够检测原图中的较大物体

```python
display_anchors(fmap_w=2, fmap_h=2, s=[0.4])
```

下面这个区域更小(fmap = 1 x 1)只有一个像素，使用了更大的锚框，便于检测原图中的更大一些的物体

```python
display_anchors(fmap_w=1, fmap_h=1, s=[0.8])
```

## 总结
通过生成多尺度锚框，成功减少了根据一张图片生成的锚框数量
