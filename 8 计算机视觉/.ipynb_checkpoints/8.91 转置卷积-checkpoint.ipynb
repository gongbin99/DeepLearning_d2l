## 转置卷积的计算方式
padding作用于output会之变小，stride很大的话会使output很大



```python
import torch
from torch import nn
from d2l import torch as d2l
```

## 转置卷积基本操作的实现

```python
def trans_conv(X, K):
    h, w = K.shape
    # 计算出输出结果的尺寸
    Y = torch.zeros((X.shape[0] + h - 1, X.shape[1] + w - 1))
    # 每个元素乘以卷积核，将这些中间结果累计起来得到结果
    for i in range(X.shape[0]):
        for j in range(X.shape[1]):
            Y[i: i + h, j: j + w] += X[i, j] * K
    return Y
```

```python
X = torch.tensor([[0.0, 1.0], 
                  [2.0, 3.0]])

K = torch.tensor([[0.0, 1.0], 
                  [2.0, 3.0]])
trans_conv(X, K)
```

调用高级API得到同样的结果

```python
X, K = X.reshape(1, 1, 2, 2), K.reshape(1, 1, 2, 2)
tconv = nn.ConvTranspose2d(1, 1, kernel_size=2, bias=False)
tconv.weight.data = K
tconv(X)
```

## 填充、步幅和多通道
与卷积不同，在转置卷积中，填充被应用于的输出（常规卷积将填充应用于输入）。 例如，当将高和宽两侧的填充数指定为1时，转置卷积的输出中将删除第一和最后的行与列。


```python
tconv = nn.ConvTranspose2d(1, 1, kernel_size=2, padding=1, 
                           bias=False)
tconv.weight.data = K
tconv(X)
```

与卷积不同，在转置卷积中，步幅被指定给中间结果在输出上的滑动幅度。当stride=2，中间结果在输出上每次滑动两步


```python
tconv = nn.ConvTranspose2d(1, 1, kernel_size=2, stride=2, 
                           bias=False)
tconv.weight.data = K
tconv(X)
```

如果我们将X代入卷积层f来输出Y=f(X)，并创建一个与具有相同的超参数、但输出通道数量是X中通道数的转置卷积层g，那么g(Y)的形状将与
X相同。 

```python
X = torch.rand(size=(1, 10, 16, 16))
conv = nn.Conv2d(10, 20, kernel_size=5, padding=2, stride=3)
tconv = nn.ConvTranspose2d(20, 10, kernel_size=5, padding=2, 
                           stride=3)
tconv(conv(X)).shape == X.shape
```

## 转置卷积与‘转置’的联系

可以使用矩阵乘法来实现卷积的互相关运算

```python
X = torch.arange(9.0).reshape(3, 3) # 3x3的输入
K = torch.tensor([[1.0, 2.0], [3.0, 4.0]]) # 2x2的卷积核
Y = d2l.corr2d(X, K) # 这里互相关运算不是矩阵乘法形式实现的
Y
```

我们将卷积核K重写为包含大量0的稀疏权重矩阵W

```python
def kernel2matrix(K):
    """通过卷积核得到权重矩阵"""
    # X拉成向量后是9维的，根据公式可知输出Y是2x2的矩阵，是4维的，因此权重矩阵是4x9
    # k中存放了W包含的元素值 k=[1, 2, 0, 3, 4]
    k, W = torch.zeros(5), torch.zeros((4, 9))
    k[:2], k[3:5] = K[0, :], K[1, :]
    W[0, :5], W[1, 1:6], W[2, 3:8], W[3, 4:] = k, k, k, k
    return W

W = kernel2matrix(K)
W
```

```python
Y == (W @ X.reshape(-1)).reshape(2, 2)
```

同样，我们可以使用矩阵乘法来实现转置卷积的计算, 其中使用的权重矩阵就是常规卷积矩阵乘法实现中使用的权重矩阵W的转置！！这就是转置卷积，‘转置’的由来

```python
Z = trans_conv(Y, K)
Z == (W.T @ Y.reshape(-1)).reshape(-1, 3)
```

转置卷积层能够交换卷积层的正向传播函数和反向传播函数：它的正向传播和反向传播函数将输入向量分别与W和W.T相乘。
