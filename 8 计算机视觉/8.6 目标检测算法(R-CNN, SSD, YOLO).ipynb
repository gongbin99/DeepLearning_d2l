## R-CNN 区域卷积神经网络


### 兴趣区域池化层 RoI （R-CNN中的一个关键层）
使不同大小的锚框都能变成同样的大小，做成一个小批量


## Fast RCNN
Fast RCNN不再像R-CNN那样使用卷积层对每个锚框进行特征抽取，而是对整个图片进行一次特征抽取，在抽取出的feature map上找出锚框位置，将RoI Pooling作用于每个锚框得到相同的尺寸，再对每个锚框抽feature（展平成向量）送入全连接层做预测


## Faster R-CNN
小网络Rpn粗糙地返回一些比较好的锚框，再由大网络进行精细的预测


## Mask R-CNN
RoI align作用于锚框时于RoI pooling不同，其对一个像素的不同部分加权，达到真的切开一个像素的效果，为了在像素级别预测的时候不要发生太多的错位


## Faster R-CNN的性能
Faster RCNN 速度慢但是精度可以做的很高，刷榜，paper用的多


## 总结
无人车中Mask R-CNN用的多（其对每个像素都进行标号）


## SSD 单发多框检测
SSD是对一张图片的每个像素生成相同数量的锚框


底层的feature map生成的锚框识别小物体，上层识别大物体（卷积越深通常我们让通道数变多，feature map尺寸减半，与感受野大小一个道理，越深的网络中一个像素看到的内容更大）


## SSD的性能
SSD速度快，但是精度比较低（因为SSD的作者做完就走了），SSD实现简单，能够用python实现
