## BERT(NLP里的迁移学习)
迁移学习：使用在大数据集上训练好的模型对我们自己小数据集的样本进行特征抽取<br>
word2vec模型：用一个词预测周围的词，或用周围的词预测中间的词（填空）


## BERT就是只有编码器的Transformer


## 创新点：对输入的修改
在Transformer中输入样本也是句子对，源句子输入到Encoder，目标句子输入到Decoder。但BERT只有Encoder，他将两个句子拼起来放到Encoder中


## 预训练任务1：带掩码的语言模型
BERT要做个通用的任务，因为他是预训练模型<br>
文本里面最通用的任务：语言模型（根据一些词预测下一个词）<br>
BERT不能直接做语言模型，因为Encoder(attention机制)的输出是看了整个句子的内容，所以BERT做了个修改——<b>带mask的语言模型</b>相当于在做完形填空，看双向信息没问题

使得模型不要看到\<mask\>就去预测
