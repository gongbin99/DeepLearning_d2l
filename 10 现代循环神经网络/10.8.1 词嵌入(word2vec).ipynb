## 独热编码其实并不是好的选择
前面我们将样本词元通过转换为独热编码后作为网络的输入X，虽然独热向量很容易构建，但它们通常不是一个好的选择.<br>
<b>主要原因: </b>不能准确表达不同词之间的相似度，比如我们经常使用的“余弦相似度”，然而对于独热编码表示的向量x, y结果总为0；也就是说用独热编码表示词的情况，是假设了所有词之间没有关系，看不出语义资讯。
